```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# task 1
Create a Line Plot with Different Colors

•	Plot two lines on the same graph using different colors.

•	Example: Use blue for the first line and green for the second line.



```python
x = [1,2,3,4,5]
y = [5,3,3,6,7]
z = [3,4,7,3,8]

plt.plot(x,y, color="blue", linestyle= '--', marker= '^')
plt.plot(x,z, color="green", linestyle= '-', marker= 'D')
```

# task 5
Customize Plot Appearance

•	Adjust the line width, marker size, and add labels to the axes.

•	Example: Increase the line width to 2, marker size to 10, and add labels and a title.


```python
x = [1,2,3,4,5]
a = [5,3,3,6,7]
b = [3,4,7,3,8]


plt.plot(x,a, color="red", linestyle= '-', marker= 'o', linewidth='2', markersize='10')
plt.plot(x,b, color="green", linestyle= '--', marker= 's', linewidth='2', markersize='10')

plt.title("this is title")
plt.xlabel("this is x label")
plt.ylabel("this is y label")
```

# task 6
6. Add a Legend to Your Plot

•	Create a plot with multiple lines and add a legend to distinguish between them.

•	Example: Label each line with a different color and style in the legend.


```python
x = [1,2,3,4,5]
a = [5,3,3,6,7]
b = [3,4,7,3,8]
c = [1,2,4,6,2]

plt.plot(x,a, color="red", linestyle= '-', marker= 'o', label ="red line")
plt.plot(x,b, color="green", linestyle= '--', marker= 's', label ="green line")
plt.plot(x,c, color="blue", linestyle= ':', marker= 'D', label = "blue line")
plt.legend()

plt.title("this is title")
plt.xlabel("this is x label")
plt.ylabel("this is y label")
```

# task 7

```python
x = [1,2,3,4,5]
a = [5,3,3,6,7]
b = [3,4,7,3,8]

plt.plot(x,a, color="red", linestyle= '-', marker= 'o', label ="red line")
plt.plot(x,b, color="green", linestyle= '--', marker= 's', label ="green line")
plt.legend()

plt.title("this is title")
plt.xlabel("this is x label")
plt.ylabel("this is y label")
```

# task 8
8. Plot Histogram with Custom Colors

•	Create a histogram with different colors for different bins or datasets.


```python
a = [5,3,3,6,7,4,5,3,2,2,4,8,6,0,5,4,2,1,2,3,3,9]
b = [3,4,7,3,8,2,3,5,6,7,5,4,6,8,9,7,8,7,6,6,0,1]
c = [1,2,5,6,2,0,8,0,2,1,1,2,9,2,1,2,3,4,4,4,4,7]

plt.hist(a, bins = 10, color="red", label ="red line", alpha= 0.5)
plt.hist(b, bins = 10, color="green", label ="green line", alpha= 0.5)
plt.hist(c, bins = 10, color="blue", label ="green line", alpha= 0.5)

plt.legend()

plt.title("this is title")
plt.xlabel("this is x label")
plt.ylabel("this is y label")
```
