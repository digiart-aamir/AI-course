```python
# important libraries
import pandas as pd
import numpy as np
```

```python
data1 = {"key": ['A','B','C','D'],
         'value':[1,2,3,4]}

data2 = {"key": ['B','D','E','F'],
        'value': [5,6,7,8]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
```

```python
# inner join

inner = pd.merge(df1,df2,
                on='key',
                how='inner')

inner
```

```python
# outer join

outer = pd.merge(df1,df2,
                on='key',
                how='outer')

outer
```

```python
# left join

left = pd.merge(df1,df2,
                on='key',
                how='left')

left
```

```python
# right join

right = pd.merge(df1,df2,
                on='key',
                how='right')

right
```

```python
# simple concatination
concat = pd.concat([df1,df2])
concat
```

```python
# concatination with axis 1 
concat = pd.concat([df1,df2], axis=1)
concat
```
